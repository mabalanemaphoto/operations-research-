# Building a model using PuLP

PuLP is an open source solver, linear and mixed integer programming modeler in Python. We use it to create and solve MILP optimisation.problems 

```python
# Install the PuLP library for linear programming
#pip install pulp

from pulp import *
```

$$
\begin{array}{rll}
    \text{max} & x+y+2z\\
    \text{s.t} & x+2y+4z \le 4 \\ & 
    x+y \ge 1 \\[10pt] &
    x,y,z \ge 0 
\end{array}
$$

```python
# Create the 'prob' variable to contain the problem data
LP_prob = LpProblem("Linear Programming_Problem", LpMinimize)

# Create decision variables
x = LpVariable('x', lowBound=0)
y = LpVariable('y', lowBound=0) 
z = LpVariable('z', lowBound=0)

# Objective function
LP_prob += x+y+2*z, "Total Cost"

# Constraints
LP_prob += x+ 2*y +4*z <= 4 , "Constraint 1"
LP_prob += x+y>= 1 , "Constraint 2"

# The problem data is written to an .lp file
LP_prob.writeLP("LP_Problem.lp")


# The problem is solved using PuLP's choice of Solver
LP_prob.solve()


# The status of the solution is printed to the screen
print("Status:", LpStatus[LP_prob.status])

# Each of the variables is printed with it's resolved optimum value
for v in LP_prob.variables():
    print(v.name,"=", v.varValue)

```

```python
# Optimal value of the objective function
print("Total Cost = ", value(LP_prob.objective))
```
